```python
import numpy as np
import pinocchio as pin
from comparer import Comparer
from IPython.display import clear_output
```

```python
# slope_list = ["00", "05", "10", "15", "20", "25", "30"]
base_list = ["500", "525", "550", "575", "600", "625", "650"]
# velocity_list = ["10", "15", "20", "25", "30", "35", "40", "45", "50"]
slope_list = ["00"]
# base_list = ["575"]
velocity_list = ["20"]

comp = Comparer(
    slope_list=[round(float(slope)*1e-4, 5) for slope in slope_list],
    base_list=[round(float(base)*1e-3, 5) for base in base_list],
    velocity_list=[round(float(velocity)*1e-2, 5) for velocity in velocity_list]
)
clear_output(wait=False)
```

```python
comp.logs = {}
comp.names = []
default_slope = "00"
default_base = "575"
default_velocity = "20"
# for slope in slope_list:
#     path = f"../results/walk_slope_{slope}_base_{default_base}_vel_{default_velocity}"
#     print(f"{path}/closed.npy")
#     comp.add_log(
#         path=f"{path}/closed.npy",
#         robot_type="closed",
#         name=f"slope_{slope}_result",
#         s=round(float(slope)*1e-4, 5),
#         b=round(float(default_base)*1e-3, 5),
#         v=round(float(default_velocity)*1e-2, 5)
#         )
#     comp.add_log(
#         path=f"{path}/closed_ws.npy",
#         robot_type="closed",
#         name=f"slope_{slope}_warmstart",
#         s=round(float(slope)*1e-4, 5),
#         b=round(float(default_base)*1e-3, 5),
#         v=round(float(default_velocity)*1e-2, 5)
#         )
for base in base_list:
    path = f"../results/walk_slope_{default_slope}_base_{base}_vel_{default_velocity}"
    comp.add_log(
        path=f"{path}/closed.npy",
        robot_type="closed",
        name=f"base_{base}_result",
        s=round(float(default_slope)*1e-4, 5),
        b=round(float(base)*1e-3, 5),
        v=round(float(default_velocity)*1e-2, 5)
        )
    comp.add_log(
        path=f"{path}/closed_ws.npy",
        robot_type="closed",
        name=f"base_{base}_ws",
        s=round(float(default_slope)*1e-4, 5),
        b=round(float(base)*1e-3, 5),
        v=round(float(default_velocity)*1e-2, 5)
        )
    comp.add_log(
        path=f"{path}/closed_warmstarted.npy",
        robot_type="closed",
        name=f"base_{base}_warmstarted",
        s=round(float(default_slope)*1e-4, 5),
        b=round(float(base)*1e-3, 5),
        v=round(float(default_velocity)*1e-2, 5)
        )
    comp.add_log(
        path=f"{path}/closed_warmstarted_ws.npy",
        robot_type="closed",
        name=f"base_{base}_warmstarted_ws",
        s=round(float(default_slope)*1e-4, 5),
        b=round(float(base)*1e-3, 5),
        v=round(float(default_velocity)*1e-2, 5)
        )
    comp.add_log(
        path=f"{path}/closed_warmstarted_warmstarted.npy",
        robot_type="closed",
        name=f"base_{base}_warmstarted_warmstarted",
        s=round(float(default_slope)*1e-4, 5),
        b=round(float(base)*1e-3, 5),
        v=round(float(default_velocity)*1e-2, 5)
        )

# for velocity in velocity_list:
#     path = f"../results/walk_slope_{default_slope}_base_{default_base}_vel_{velocity}"
#     comp.add_log(
#         path=f"{path}/closed.npy",
#         robot_type="closed",
#         name=f"velocity_{velocity}_result",
#         s=round(float(default_slope)*1e-4, 5),
#         b=round(float(default_base)*1e-3, 5),
#         v=round(float(velocity)*1e-2, 5)
#         )
#     comp.add_log(
#         path=f"{path}/closed_ws.npy",
#         robot_type="closed",
#         name=f"velocity_{velocity}_warmstart",
#         s=round(float(default_slope)*1e-4, 5),
#         b=round(float(default_base)*1e-3, 5),
#         v=round(float(velocity)*1e-2, 5)
#         )

comp.printNames()
```
