```python
import pinocchio as pin
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib

sns.set_palette("colorblind")
matplotlib.rcParams.update({'font.size': 20})
SMALL_SIZE = 10
MEDIUM_SIZE = 18
BIGGER_SIZE = 20

plt.rc('font', size=BIGGER_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
# plt.rc('axes', lengendsize=BIGGER_SIZE)    # fontsize of the legend
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
# Load the data
ja_left_knee = np.load("../results/ja_knee_left.npy")
ja_right_knee = np.load("../results/ja_knee_right.npy")
```

```python
ja_non_null = [j for j in ja_left_knee if j != 0 and j < 2.5]
print(ja_non_null)
print(len(ja_non_null))
mid = len(ja_non_null) // 2
print(mid)
diff = len(ja_non_null) - mid
list_x = np.arange(-diff+1, diff)
print(len(list_x))
```

```python
plt.figure(figsize=(7, 5))
# plt.plot(ja_left_knee, label="Left knee")
plt.axhline(1, xmin=0, xmax=1, color='red', linestyle='--', linewidth=2)
plt.plot(list_x, ja_non_null, label="Left knee", linewidth=3)
# plt.plot(ja_right_knee, label="Right knee")
plt.xlabel("Knee joint angle [deg]")
plt.ylabel("Reduction ratio")
# plt.legend()
plt.ylim(0, 4)
plt.xlim(-25, 25)
plt.grid(axis="y")
plt.tight_layout()
plt.savefig("/home/ldematteis/Images/ICRA_2025/results/joint_angle_knee.svg", format="svg")
plt.show()
```
